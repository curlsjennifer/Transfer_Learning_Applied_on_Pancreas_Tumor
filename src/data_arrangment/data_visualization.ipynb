```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
path = '/home/d/pancreas/new_box_data/PT6'
```

```python
image = np.load(path + '/ctscan.npy')
pancreas = np.load(path + '/pancreas.npy')
lesion = np.load(path + '/lesion.npy')
```

```python
print(image.shape, pancreas.shape, lesion.shape)
```

```python
def plot_fig(image, pancreas, lesion, z):
    plt.figure(figsize=(16, 10))
    plt.subplot(1, 3, 1)
    plt.imshow(image[:,:,z].T, cmap='gray')
    plt.subplot(1, 3, 2)
    plt.imshow(pancreas[:,:,z].T)    
    plt.subplot(1, 3, 3)
    plt.imshow(lesion[:,:,z].T)    
    plt.show()
    
def plot_merge(image, pancreas, lesion, z):
    plt.figure(figsize=(16, 10))
    plt.subplot(1, 2, 1)
    plt.imshow(image[:,:,z].T+pancreas[:,:,z].T)
    plt.subplot(1, 2, 2)
    plt.imshow(image[:,:,z].T+lesion[:,:,z].T)       
    plt.show()
```

```python
ww = 300
wl = 100
image = np.clip(image, wl-ww/2, wl+ww/2)
image -= image.min()
image /= (image.max() - image.min())
```

```python
z = 50
```

```python
plot_fig(image, pancreas, lesion, z)
```

```python
plot_merge(image, pancreas, lesion, z)
```
